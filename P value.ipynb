# P Value

*" p-value is the probability of obtaining a result atleast as extreme as the observed result of the test, assuming null hypothesis."*

  Nothing in this world is perfect. Everything is perfect to an extend. P value is the measure how significant some thing is. A Milkman would provide you milk throughout the year. He may be absent once or more. You would be annoyed if he makes a significant absence from his work. When you buy one kg of onion, we cannot make an exact 1 kg it may be a little high or low. But how little is important! Thats all about the need to test the significance.
  
  P-value is the probability that this imperfection is by chance. If you get 999 g of onion in place of 1 kg, there is more probability that this is by random choice. But when you get 900g it is very less likely that it is by random chance/mistake.
  
# P test

P-test is a hypothesis test. We would set the null hypothesis and would reject it if could not get the result with the test sample. Null hypothesis often the common belief or the claims made. Lets say a company claims that there mobile has long lasting battery that is more than 24 hours. This can be set as the null hypothesis and we work with an aim to disprove this! We would set another hypothesis, which can go with when the null hypothesis is rejected. This is called alternate hypothesis as this would be the best alternative to the null hypothesis.

Step 1:
  We cannot test take all the mobiles from the company for 24 hours test! We have to choose randomly the samples to be tested.
  
Step 2:
  Obtain the test statistics from the sample mean and standard deviation. 
   test statistics, \begin{align} z = \frac{\bar X - \mu}{\sigma_s} \end{align}, where $\sigma_s$ is the  sample standard deviation.
   
Step 3: 
 Now p value is the probability of obtaining a value less extreme than the test statistics.
  

# Example
  Let us consider a company models 10000 mobiles of a model. We will perform p-test on the claim made by the company that the battery last for a day without charging. Here we assume a standard deviation of 1.5 hours. The distribution looks as in the figure. The data is standardized, so that the mean is 0 and standard deviation is 1.  

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
battery_time=np.random.normal(24,1.5,10000)
plt.hist(battery_time,40)
plt.xlabel("battery_time(in hrs)")
plt.ylabel("Count")
plt.title("Battery lasting time")
plt.show()
```

```python
std_batt=(battery_time-np.mean(battery_time))/np.std(battery_time)
plt.hist(std_batt,40)
plt.xlabel("battery_time(standardized)")
plt.ylabel("Count")
plt.title("Standardized normal Battery lasting time")
plt.show()
```

# TEST:
    Let us suppose we get 1000 mobiles and perform a test. Test mean was found to be 20 hours, with standard deviation of 2 hours. We will now perform a P-test for this case now!

```python
test_mean = 20
test_std = 2
print("Test mean:",test_mean,"\nTest Standard deviation:",test_std)
```

     Test statistics was found to be -2. When we plot this in the distribution, we found that the area of the curve to the left is equal to 0.0228. That is 2.28% probability of having such mobiles when assuming the null hypothesis. We have found the significance level and it is very low. 
 
    What is the minimum value you can expect to be significant? Generally this significance level is assumed to be 5% or 0.05. Since here we have 0.0228, it is insignificant! We have to reject the null hypothesis. 

```python
test_statistics = (20-24)/2
print("Normalized test mean:",test_statistics)
```

```python
import scipy.stats as st
st.norm.cdf(-2.0) #probability a mobile with less than 20 hrs battery time
```

```python
std_batt=(battery_time-np.mean(battery_time))/np.std(battery_time)
plt.hist(std_batt,40)
plt.vlines(-2,0,900,colors='red')
plt.xlabel("battery_time(standardized)")
plt.ylabel("Count")
plt.title("Left tail p_test")
plt.show()
```

    Area of the entire curve is 1. Area of the curve to the left of red line(test statistics) is 0.0228
    
    Case 2: In case in the test we got a mean of 23 hours and a standard deviation of one hour. How will our p-test turn out to be?

```python
test_stat=(23-24)/1
print("p value is found to be :",st.norm.cdf(test_stat) )
```

```python
std_batt=(battery_time-np.mean(battery_time))/np.std(battery_time)
plt.hist(std_batt,40)
plt.vlines(-1,0,800,colors='red')
plt.xlabel("battery_time(standardized)")
plt.ylabel("Count")
plt.title("Left tail p_test")
plt.show()
```

    Again p-value of 0.1587 implies that there is a probability of 15.87% to get such samples assuming the null hypothesis! And this case the claim is significant and we cannot reject the null hypothesis.
    
Note: We could not reject null hypothesis, does not necessarily mean that we can accept it...
